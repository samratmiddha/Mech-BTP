```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_viscoelastic_results(df):
    plt.figure(figsize=(15, 4))

    plt.subplot(1, 3, 1)
    plt.plot(df['step'], df['p'])
    plt.xlabel('Timestep')
    plt.ylabel('p')
    plt.title('p vs Timestep')

    plt.subplot(1, 3, 2)
    plt.plot(df['step'], df['strain'])
    plt.xlabel('Timestep')
    plt.ylabel('Strain')
    plt.title('Strain vs Timestep')

    plt.subplot(1, 3, 3)
    plt.plot(df['step'], df['p_over_p_eq'])
    plt.xlabel('Timestep')
    plt.ylabel('p/p_eq')
    plt.title('p/p_eq vs Timestep')

    plt.tight_layout()
    plt.show()
```

```python
df = pd.read_csv('output.txt')
plot_viscoelastic_results(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File names and corresponding relaxation times
files = [
    ('output3.txt', 0.108),
    ('output4.txt', 0.158),
    ('output5.txt', 0.208),
    ('output6.txt', 0.258),
    ('output7.txt', 0.308),
]
labels = [
    "τ₁ = 0.108 s",
    "τ₁ = 0.158 s",
    "τ₁ = 0.208 s",
    "τ₁ = 0.258 s",
    "τ₁ = 0.308 s",
]

plt.figure(figsize=(12, 5))

# Plot p/p_eq
plt.subplot(1, 2, 1)
for (fname, tau), label in zip(files, labels):
    df = pd.read_csv(fname)
    plt.plot(df['step'], df['p_over_p_eq'], label=label)
plt.xlabel('Timestep')
plt.ylabel('p/p_eq')
plt.title('p/p_eq vs Timestep')
plt.legend()

# Plot p
plt.subplot(1, 2, 2)
for (fname, tau), label in zip(files, labels):
    df = pd.read_csv(fname)
    plt.plot(df['step'], df['p'], label=label)
plt.xlabel('Timestep')
plt.ylabel('p')
plt.title('p vs Timestep')
plt.legend()

plt.suptitle('Variation with relaxation time of first Maxwell element\nTotal running time: 1.062 sec')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
