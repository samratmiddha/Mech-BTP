```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_stress_strain_vs_time(df):
    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    plt.plot(df['time'], df['strain_xx'])
    plt.xlabel('Time (s)')
    plt.ylabel('Strain_xx')
    plt.title('Strain vs Time')

    plt.subplot(1, 2, 2)
    plt.plot(df['time'], df['stress_xx'])
    plt.xlabel('Time (s)')
    plt.ylabel('Stress_xx')
    plt.title('Stress vs Time')

    plt.tight_layout()
    plt.show()
```

```python
df= pd.read_csv('output.txt')
plot_stress_strain_vs_time(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File names and corresponding strain rates
files = [
    ('output4.txt', 0.6),
    ('output5.txt', 1.0),
    ('output6.txt', 1.5),
    ('output7.txt', 0.75),
    ('output8.txt', 0.5),
]
labels = [
    "Strain rate = 0.6 /s",
    "Strain rate = 1.0 /s",
    "Strain rate = 1.5 /s",
    "Strain rate = 0.75 /s",
    "Strain rate = 0.5 /s",
]

plt.figure(figsize=(14, 5))

# Plot stress vs time for each strain rate
plt.subplot(1, 2, 2)
for (fname, rate), label in zip(files, labels):
    df = pd.read_csv(fname)
    plt.plot(df['time'], df['stress_xx'], label=label)
plt.xlabel('Time (s)')
plt.ylabel('Stress_xx')
plt.title('Stress vs Time')
plt.legend()

# Plot strain vs time for each strain rate
plt.subplot(1, 2, 1)
for (fname, rate), label in zip(files, labels):
    df = pd.read_csv(fname)
    plt.plot(df['time'], df['strain_xx'], label=label)
plt.xlabel('Time (s)')
plt.ylabel('Strain_xx')
plt.title('Strain vs Time')
plt.legend()

plt.suptitle('Variation with change in strain rate (_yield constant)')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File names and corresponding sigma_yield values
files = [
    ('output9.txt', 5),
    ('output10.txt', 10),
    ('output11.txt', 15),
    ('output12.txt', 20),
    ('output13.txt', 25),
    ('output14.txt', 30),
    ('output15.txt', 35),
    ('output16.txt', 40),
]
labels = [
    "_yield = 5 MPa",
    "_yield = 10 MPa",
    "_yield = 15 MPa",
    "_yield = 20 MPa",
    "_yield = 25 MPa",
    "_yield = 30 MPa",
    "_yield = 35 MPa",
    "_yield = 40 MPa",
]

# Full range plots
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
for (fname, sy), label in zip(files, labels):
    df = pd.read_csv(fname)
    plt.plot(df['time'], df['strain_xx'], label=label)
plt.xlabel('Time (s)')
plt.ylabel('Strain_xx')
plt.title('Strain vs Time')
plt.legend()

plt.subplot(1, 2, 2)
for (fname, sy), label in zip(files, labels):
    df = pd.read_csv(fname)
    plt.plot(df['time'], df['stress_xx'], label=label)
plt.xlabel('Time (s)')
plt.ylabel('Stress_xx')
plt.title('Stress vs Time')
plt.legend()
plt.suptitle('Variation with change in _yield (full range)')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Zoomed plots till 0.025 sec
plt.figure(figsize=(14, 10))
plt.subplot(1, 2, 1)
for (fname, sy), label in zip(files, labels):
    df = pd.read_csv(fname)
    df_zoom = df[df['time'] <= 0.025]
    plt.plot(df_zoom['time'], df_zoom['strain_xx'], label=label)
plt.xlabel('Time (s)')
plt.ylabel('Strain_xx')
plt.title('Strain vs Time (Zoomed)')
plt.legend()

plt.subplot(1, 2, 2)
for (fname, sy), label in zip(files, labels):
    df = pd.read_csv(fname)
    df_zoom = df[df['time'] <= 0.025]
    plt.plot(df_zoom['time'], df_zoom['stress_xx'], label=label)
plt.xlabel('Time (s)')
plt.ylabel('Stress_xx')
plt.title('Stress vs Time (Zoomed)')
plt.legend()
plt.suptitle('Variation with change in _yield (zoomed till 0.010 s)')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# File names and corresponding E_infinity values
files = [
    ('output25.txt', 3500),
    ('output24.txt', 3000), 
    ('output21.txt', 2500),
    ('output22.txt', 2000),
    ('output23.txt', 1500),
]
labels = [
    "E_infinity = 3500 MPa",
    "E_infinity = 3000 MPa",
    "E_infinity = 2500 MPa",
    "E_infinity = 2000 MPa",
    "E_infinity = 1500 MPa",
]

# Plot combined graphs
plt.figure(figsize=(14, 5))

# Strain vs Time
plt.subplot(1, 2, 1)
for fname, e_inf in files:
    df = pd.read_csv(fname)
    plt.plot(df['time'], df['strain_xx'], label=f"E_infinity = {e_inf} MPa")
plt.xlabel('Time (s)')
plt.ylabel('Strain_xx')
plt.title('Strain vs Time')
plt.legend()

# Stress vs Time
plt.subplot(1, 2, 2)
for fname, e_inf in files:
    df = pd.read_csv(fname)
    plt.plot(df['time'], df['stress_xx'], label=f"E_infinity = {e_inf} MPa")
plt.xlabel('Time (s)')
plt.ylabel('Stress_xx')
plt.title('Stress vs Time')
plt.legend()

plt.suptitle('Variation of Stress with Change in E_infinity')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
